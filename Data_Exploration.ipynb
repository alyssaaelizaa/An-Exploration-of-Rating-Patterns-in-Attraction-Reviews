```python
import pandas as pd
import matplotlib.pyplot as plt

# reading csv file using pandas
google_ratings = pd.read_csv("google_review_ratings.csv")
print(google_ratings)
```

```python
# displaying data types of each column
google_ratings.dtypes
```

```python
# summary statistics of google ratings file
google_ratings.mean()
```

```python
# statistics plot of minimum value

numerical = google_ratings.select_dtypes('float64')
min_plot = numerical.min()
min_plot.plot.bar()
plt.ylabel('Minimum Value')
plt.xlabel('Numerical Columns')
plt.title('Minimum Values of Numerical Columns')
plt.show()
```

```python
# statistics plot of maximum value

numerical = google_ratings.select_dtypes('float64')
max_plot = numerical.max()
max_plot.plot.bar()
plt.ylabel('Maximum Value')
plt.xlabel('Numerical Columns')
plt.title('Maximum Values of Numerical Columns')
plt.show()
```

```python
# statistics plot of median value

numerical = google_ratings.select_dtypes('float64')
median_plot = numerical.median()
median_plot.plot.bar()
plt.ylabel('Median Value')
plt.xlabel('Numerical Columns')
plt.title('Median Values of Numerical Columns')
plt.show()
```

```python
# statistics plot of standard deviation

numerical = google_ratings.select_dtypes('float64')
std_plot = numerical.std()
std_plot.plot.bar()
plt.ylabel('Standard Deviation')
plt.xlabel('Numerical Columns')
plt.title('Standard Deviation of Numerical Columns')
plt.show()
```

```python
# statistics plot of mean deviation

numerical = google_ratings.select_dtypes('float64')
mean_plot = numerical.mean()
mean_plot.plot.bar()
plt.ylabel('Mean')
plt.xlabel('Numerical Columns')
plt.title('Mean of Numerical Columns')
plt.show()
```

```python
google_ratings.describe()
```

```python
# box plot displaying two parameters

restaurants = 'Category 9'
parks = 'Category 4'

compare = google_ratings[[restaurants, parks]]
compare.boxplot()

plt.xlabel("Categories")
plt.ylabel("Ratings")
plt.title("Box Plot Distribution of Restaurant vs. Park Ratings")
```
